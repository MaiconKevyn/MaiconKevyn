```python
import math
```

```python
x = float(input ('Digite o valor x:'))
y = float(input ('Digite o valor y:'))

r = math.sqrt( x * x + y * y )
theta = math.atan(y/x)
theta = 180*theta/math.pi #theta em radiano
print('As coordenadas polares são: r = %0.2f,theta = %0.2f'%(r, theta))
```
